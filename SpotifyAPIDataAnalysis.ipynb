```python
import requests

CLIENT_ID = 'Your Client ID'
CLIENT_SECRET = 'Your Client Secret'

AUTH_URL = 'https://accounts.spotify.com/api/token'
```

```python
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
client_credentials_manager = SpotifyClientCredentials(client_id=CLIENT_ID, client_secret=CLIENT_SECRET)
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)
```

Transfer the data from the json in lists

```python
artist_name = []
track_name = []
track_popularity = []
artist_id = []
track_id = []
for i in range(0,1000):
    track_results = sp.search(q='year:2022', type='track', limit=50,offset=i)
    for i, t in enumerate(track_results['tracks']['items']):
        artist_name.append(t['artists'][0]['name'])
        artist_id.append(t['artists'][0]['id'])
        track_name.append(t['name'])
        track_id.append(t['id'])
        track_popularity.append(t['popularity'])
```

Create a dataframe with the lists as columns

```python
import pandas as pd
track_df = pd.DataFrame({'artist_name' : artist_name, 'track_name' : track_name, 'track_id' : track_id, 'track_popularity' : track_popularity, 'artist_id' : artist_id})
print(track_df.shape)
track_df_pop = track_df.sort_values("track_popularity", ascending=False)
track_df_pop.head(10)
```

There seem to be an awful lot of duplicate values for songs. We have to clean them up.

```python
track_df_pop_no_d = track_df_pop.drop_duplicates("track_id")
print(track_df_pop_no_d.shape)
track_df_pop_no_d.head(10)
```

# Nice. Now we have the top 10 songs of 2022 so far.

Let's see who the most popular artists are.


```python
artist_popularity = []
artist_genres = []
artist_followers = []
for a_id in track_df_pop_no_d.artist_id:
  artist = sp.artist(a_id)
  artist_popularity.append(artist['popularity'])
  artist_genres.append(artist['genres'])
  artist_followers.append(artist['followers']['total'])
```

```python
track_df_pop_no_d = track_df_pop_no_d.assign(artist_popularity=artist_popularity, artist_genres=artist_genres, artist_followers=artist_followers)
track_df_pop_no_d.head(10)
```

```python
by_art_pop = pd.DataFrame(track_df_pop_no_d.sort_values(by=['artist_popularity'], ascending=False)[['artist_name','artist_popularity','artist_followers','artist_genres']])
by_art_pop.astype(str).drop_duplicates().head(10)
```

# These are the top artists of 2022 so far

```python
track_features = []
for t_id in track_df_pop_no_d['track_id']:
  af = sp.audio_features(t_id)
  track_features.append(af)
```

```python
tf_df = pd.DataFrame(columns = ['danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'type', 'id', 'uri', 'track_href', 'analysis_url', 'duration_ms'])
```

```python
for item in track_features:
  for feat in item:
    tf_df = tf_df.append(feat, ignore_index=True)
```

```python
tf_df.head()
```

```python
tf_df['duration_ms'] = pd.to_numeric(tf_df['duration_ms'])
cols_to_drop2 = ['key','mode','type', 'uri','track_href','analysis_url']
tf_df = tf_df.drop(columns=cols_to_drop2)
print(track_df_pop_no_d.info())
print(tf_df.info())
```

```python
track_df_pop_no_d['artist_name'] = track_df_pop_no_d['artist_name'].astype("string")
track_df_pop_no_d['track_name'] = track_df_pop_no_d['track_name'].astype("string")
track_df_pop_no_d['track_id'] = track_df_pop_no_d['track_id'].astype("string")
track_df_pop_no_d['artist_id'] = track_df_pop_no_d['artist_id'].astype("string")
tf_df['instrumentalness'] = pd.to_numeric(tf_df['instrumentalness'])
print(track_df_pop_no_d.info())
print(tf_df.info())
```

```python
tf_df = tf_df.rename(columns = {'id':'track_id'})
```

```python
top_track_info = pd.merge(
    track_df_pop_no_d.reset_index(),
    tf_df.reset_index(),
    on = ["track_id"]
)
```

# Now we have the stats for the top 10 songs of 2022

Now I want to know the stats of one of my favourite songs: "Lateralus" by Tool. Because it came out in 2001 it most definitely is not in the top songs of 2022. So we will have to send a new request to the Spotify API

```python
lateralus_uri = "spotify:track:7tvuLLroI0n6uYBWuFig5d"
lat_features = sp.audio_features(lateralus_uri)
lat_features
```

Okay. Looks interesting. Apparently these are the audio features of my favourite song. I wonder what other songs Spotify can recommend to me based on these values.

```python
rec = sp.recommendations(seed_tracks=[lateralus_uri], limit=10)
for track in rec['tracks']:
  print(track['artists'][0]['name'],"-", track['name'])
```

That's great, but soon I will travel to Korea. Naturally, I want to experience the culture and I am curious if Spotify can recommend any tracks in the kpop genre that are similar to Lateralus.

Sadly, I could not get the API to return such values. The reason being, that you can only include a seed_genre as kpop, but that does not guarantee kpop output. Makes sense. So I tried limiting the otput to show only tracks that have their genre set as kpop. But for some reason the output of the API does not include information on the genre.

```python
rec = sp.recommendations(seed_tracks=[lateralus_uri], seed_genres=["kpop"], limit=10)
for track in rec['tracks']:
    print(track['artists'][0]['name'],"-", track['name'])
```
